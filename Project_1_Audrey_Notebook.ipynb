```python
# import dependencies
import pandas as pd
import os
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
# Read CSV and display DataFrame
df = pd.read_csv("spotify_dataset.csv")
df.head()
```

```python
# Display DataFrame for the artist "The Weeknd"
artist_df = (df[df['Artist'] == 'The Weeknd']).copy()
artist_df
```

```python
# Display number of followers for The Weeknd
followers = artist_df['Artist Followers'].max()
print(followers)
```

```python
# Display how many songs The Weeknd has in top 200
song_count = artist_df["Song ID"].count()
song_count
```

```python
# Display the total number of times songs have charted for The Weeknd
total_times_charted = artist_df["Number of Times Charted"].sum()
total_times_charted
```

```python
# Display whether or not The Weeknd has had a #1 charting song
if len(artist_df.loc[artist_df["Highest Charting Position"] == 1]) > 0:
    print("The Weeknd has topped the chart")
else:
    print("The Weeknd has not topped the chart")
```

```python
# Specific song by The Weeknd that hit #1
artist_df.loc[artist_df["Highest Charting Position"] == 1,["Song Name"]]
```

```python
# Convert streams column from a string to an integer
artist_df['Streams'] = artist_df['Streams'].str.replace(',',"").astype(int)
```

```python
# Create a new DataFrame for The Weeknd showing the song name and number of streams
artist_df_streams = artist_df[['Song Name','Streams']]
artist_df_streams
```

```python
# Display total number of streams for The Weeknd
Weeknd_total_streams = artist_df_streams['Streams'].sum()
Weeknd_total_streams
```

```python
# create bar graph showing number of streams and song title
bar_graph = artist_df.groupby('Song Name')['Streams'].sum()
bar_graph.plot(kind="bar", x="Song Name", y="Streams")

plt.xlabel("Song Name")
plt.ylabel("Streams (in ten millions)")
plt.xticks(rotation=90)
plt.title("The Weeknd's Number of Streams/Song")

plt.savefig("Weeknd_bar_graph.png")
plt.show()
```
