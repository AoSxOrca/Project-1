```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
# load the dataset into a pandas dataframe
spotify_df = pd.read_csv("C:/Users/Jason/Desktop/spotify_dataset.csv")

spotify_df
```

```python
#create artist specific dataframe
ed_sheeran_df = spotify_df[spotify_df["Artist"] == "Ed Sheeran"].copy()
ed_sheeran_df
```

```python
#display how many followers artist has using max function
followers = ed_sheeran_df['Artist Followers'].max()
print(followers)
```

```python
#Display how many songs artist has in top 200
song_count = ed_sheeran_df["Song ID"].count()
song_count
```

```python
#Total number of times Ed Sheeran has made it to the top 200
times_charted = ed_sheeran_df["Number of Times Charted"].sum()
times_charted
```

```python
#Display how many streams artist has from the songs in the top 200
ed_sheeran_streams = ed_sheeran_df["Streams"].sum()
ed_sheeran_streams
```

```python
#Display all of the songs Ed Sheeran has in the top 200
ed_sheeran_streams_df = ed_sheeran_df[["Song Name", "Streams"]].copy()
ed_sheeran_streams_df.sort_values(by=["Streams"], ascending=False)
```

```python
#Display the total number of streams that Ed Sheeran has received in the top 200
ed_total_streams = ed_sheeran_streams_df["Streams"].sum()
ed_total_streams
```

```python
## display whether or not artist has had a #1 charting song
if len(ed_sheeran_df.loc[ed_sheeran_df["Highest Charting Position"] == 1]) > 0:
    print("Ed Sheeran has topped the chart")
else:
    print("Ed Sheeran has not topped the chart")
```

```python
# convert ‘Streams’ column from strings into integers
ed_sheeran_df['Streams'] = ed_sheeran_df['Streams'].str.replace(",","").astype(int)
```

```python
# create bar graph showing number of streams and song title for Ed Sheeran
bar_graph = ed_sheeran_df.groupby('Song Name')['Streams'].sum()
bar_graph.plot(kind="bar", x="Song Name", y="Streams")

plt.xlabel("Song Name")
plt.ylabel("Streams (in ten millions)")
plt.xticks(rotation = 90)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the pie chart
artists = ["Doja Cat", "Ed Sheeran", "Drake", "The Weeknd"]
streams = [133496415, 80580568, 128538281, 149088439]

# Plot the pie chart
plt.pie(streams, labels=artists, autopct="%1.1f%%")
plt.axis("equal")
plt.title("Total Number of Streams")
plt.show()
```

```python
labels = ['Doja Cat', 'Ed Sheeran', 'Drake', 'The Weeknd']
sizes = [133496415, 80580568, 128538281, 149088439]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

#draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.75,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
#Create dataframes for other artists
weeknd_count = (spotify_df[spotify_df["Artist"] == "The Weeknd"]).copy()
drake_count = (spotify_df[spotify_df["Artist"] == "Drake"]).copy()
doja_count = (spotify_df[spotify_df["Artist"] == "Doja Cat"]).copy()
```

```python
#Follower count for each artist
weeknd_followers = weeknd_count["Artist Followers"].max()
drake_followers = drake_count["Artist Followers"].max()
doja_followers = doja_count["Artist Followers"].max()

print("The Weeknd has" , weeknd_followers , "followers")
print("Drake has" , drake_followers , "followers")
print("Doja Cat has" , doja_followers , "followers")
```

```python
#Converting column data to int for each dataframe
weeknd_count["Streams"] = weeknd_count["Streams"].str.replace(",","").astype(int)
drake_count["Streams"] = drake_count["Streams"].str.replace(",","").astype(int)
doja_count["Streams"] = doja_count["Streams"].str.replace(",","").astype(int)
```

```python
#Horizontal bar chart visualizing total number of streams per artist
plt.barh(artists, streams, color=("Blue", "Orange", "Green", "Red"))
plt.title("Total number of streams per artist")
plt.ylabel("Artists")
plt.xlabel("Number of streams (in hundred millions)")
plt.savefig("totalhorizbarchart.png")
plt.show()
```

```python
#Display how many times other artists have charted the top 200
weeknd_charted = weeknd_count["Number of Times Charted"].sum()
drake_charted = drake_count["Number of Times Charted"].sum()
doja_charted = doja_count["Number of Times Charted"].sum()

print("The Weeknd has charted", weeknd_charted, "times")
print("Drake has charted", drake_charted, "times")
print("Doja Cat has charted", doja_charted, "times")
print("Ed Sheeran has charted", times_charted, "times")
```

```python
Horizontal_charted = [doja_charted, times_charted, drake_charted, weeknd_charted]
```
